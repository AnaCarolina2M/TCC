## A-10-Adult Questionnaire
- 1 - вҖңI often notice small sounds when others do not.вҖқ
- 2 - вҖңI usually concentrate more on the whole picture, rather than the small details.вҖқ
- 3 - вҖңI find it easy to do more than one thing at once.вҖқ
- 4 - вҖңIf there is an interruption, I can switch back to what I was doing very quickly.вҖқ
- 5 - вҖңI find it easy to вҖҳread between the linesвҖҷ when someone is talking to me.вҖқ
- 6 - вҖңI know how to tell if someone listening to me is getting bored.вҖқ
- 7 - вҖңWhen IвҖҷm reading a story I find it difficult to work out the charactersвҖҷ intentions.вҖқ
- 8 - вҖңI like to collect information about categories of things (e.g. types of cars, birds, trains, plants).вҖқ
- 9 - вҖңI find it easy to work out what someone is thinking or feeling just by looking at their face.вҖқ
- 10 - вҖңI find it difficult to imagine what it would be like to be someone else.вҖқ

```python
import pandas as pd
from scipy.io import arff

file = "../data/Autism-Adult-Data.arff"
data, meta = arff.loadarff(file)
df = pd.DataFrame(data)
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.decode('utf-8')
```

```python
df.head()
```

```python
def df_info(df:pd.DataFrame):
    info = df.info()
    columns = df.columns
    nulls = print("Nulls:",df.isnull().sum())
    nan = print("?_count:",df.isin(['?']).sum().sum())
    return info,columns,nulls,nan
```

```python
info = df_info(df)
info
```

```python
def quantity_of_registers(col: pd.Series):
    return len(df[col].unique())
```

```python
qtd_countries = quantity_of_registers("contry_of_res")
qtd_ethnicities = quantity_of_registers("ethnicity")
qtd_age_groups = quantity_of_registers("age_desc")
qtd_countries, qtd_ethnicities,qtd_age_groups
```
